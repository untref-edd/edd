```python
from edd.grafo import Grafo

G = Grafo()
G.agregar_arista("A", "B")
G.agregar_arista("A", "C")
G.agregar_arista("B", "D")
G.agregar_arista("B", "E")
G.agregar_arista("D", "F")
G.agregar_arista("D", "C")
G.agregar_arista("E", "C")
G.agregar_arista("C", "G")
G.agregar_arista("F", "G")
G.agregar_arista("E", "F")

G.draw(output_file="figuras/grafo-bipartito.png", nodes_group=["A", "D", "E", "G"])
G.draw(
    output_file="figuras/grafo-bipartito-ordenado.png",
    nodes_group=["A", "D", "E", "G"],
    pos={"A": (1, 1), "D": (2, 1), "E": (3, 1), "G": (4, 1), "B": (1.5, 0), "C": (2.5, 0), "F": (3.5, 0)},
)
```
